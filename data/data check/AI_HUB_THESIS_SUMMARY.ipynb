## 1. json 파일 읽기 (테스트)

### 논문

```python
with open('data/AI_HUB/AI_HUB_THESIS_SUMMARY/Training/thesis/논문요약_0206_0.json', "r", encoding='utf8') as json_file:
    load_json = json.load(json_file)
```

```python
load_json.keys()
```

```python
load_json['totalcount']
```

```python
load_json['data'][0]
```

```python
len(load_json['data'])
```

```python
sentence_list = []
summary_list = []

for idx, text in enumerate(load_json['data']):
    if idx == 0:
        entire = text['summary_entire']
        section = text['summary_section']
        
        sentence_list.append(entire[0]['orginal_text'])
        summary_list.append(entire[0]['summary_text'])
        sentence_list.append(section[0]['orginal_text'])
        summary_list.append(section[0]['summary_text'])
```

```python
sentence_list
```

```python
summary_list
```
